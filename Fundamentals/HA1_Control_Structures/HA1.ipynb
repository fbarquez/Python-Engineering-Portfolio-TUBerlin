### 2) Richtwertüberschreitung

  * Berechnen Sie für jeden Monat die Anzahl der Tage, an denen der [zulässige Richtwert](http://www.verwaltungsvorschriften-im-internet.de/bsvwvbund_26081998_IG19980826.htm) für urbane Gebiete von 63 dB überschritten wurde. 
  * Berechnen Sie deren prozentualen Anteil an der Gesamtzahl der Werktage im jeweiligen Monat. 
  * Geben Sie das Ergebnis tabellarisch aus.
  * Vermeiden Sie auch hier die wiederholte Eingabe nahezu identischer Code-Abschnitte.
  * Speichern Sie außerdem für einen automatisierten Test die Anzahl für den Monat Juni in der Variable `anzahl_juni`. (Für die anderen Monate müssen keine eigenen Variablen erstellt werden.)

Beispiel:

    Juli    :  6 Tage (28.6 %)
    August  :  9 Tage (39.1 %)
    
<!-- Werktage Juli:21, August: 23 -->

```python

# Hier eigenen Code schreiben ...


anzahl_juni = None # <- diese Variable sollte später die Anzahl an Juni-Tagen mit Richtwert-Überschreitung enthalten

# Richtwertüberschreitungen > 63 dB je Monat
richtwert = 63

print("\n--- Richtwertüberschreitung (> 63 dB) ---")
for monat, werte in LAeq.items():
    gesamt = len(werte)
    ueber = sum(wert > richtwert for wert in werte)
    prozent = (ueber / gesamt) * 100 if gesamt else 0.0
    print(f"{monat:8s}: {ueber:2d} Tage ({prozent:4.1f} %)")

    if monat == "Juni":
        anzahl_juni = ueber

assert anzahl_juni is not None
assert anzahl_juni < 10

```

```python
# Hier ist ein Plausibilitätstest
assert anzahl_juni < 10
```

**3. Durch intensive Maßnahmen zur Lärmminderung und -vermeidung könnte der Pegel um durchschnittlich 2 dB gesenkt werden.** 

3.1) _Schätzen_ Sie basierend auf den gegebenen Werten ungefähr ab, wie viele (Werk-)Tage _im Jahr_ es noch immer zu Richtwertüberschreitungen kommen würde und geben Sie das Ergebnis aus.

```python
anzahl_jahr = None # <- diese Variable sollte später die geschätzte Anzahl an Tagen im Jahr mit Richtwert-Überschreitung enthalten
# Hier eigenen Code schreiben ...

reduktion = 2
grenzwert = 63 + reduktion
ueberschreitungen_6_monate = sum(1 for werte in LAeq.values() for p in werte if p > grenzwert)
anzahl_jahr = int(round((ueberschreitungen_6_monate / 6) * 12))
print(anzahl_jahr)
assert anzahl_jahr > 40
```
